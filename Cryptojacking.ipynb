```python
import pandas as pd
import numpy as np
```

# **Read CSV from GOOGLE DRIVE**

```python
def read_csv_in_google_drive(url):
    url='https://drive.google.com/uc?id=' + url.split('/')[-2]
    data_frame= pd.read_csv(url, encoding='utf-8')
    return data_frame
```

```python
normal_sheet_url='https://drive.google.com/file/d/16L8-2Y6JfA8OMaiNFm4dX2KGk-CP4Lsz/view?usp=sharing'

normal_df = read_csv_in_google_drive(normal_sheet_url)

```

```python
normal_df.head()
```

```python
anormal_sheet_url='https://drive.google.com/file/d/1AXgrm0NvZ-oC1jsPJDiVAde0bn2V03uq/view?usp=share_link'

anormal_df = read_csv_in_google_drive(anormal_sheet_url)

```

```python
anormal_df.head()
```

```python
print(anormal_df.shape)
print(normal_df.shape)
```

# **MERGE THE DATASETS**

```python
anormal_df['flag'] = 1 # 1 for problem
normal_df['flag'] = 0 # 0 for no problem
df = pd.concat([anormal_df, normal_df], axis = 0) # adding the records
```

```python
df.head()
```

# **CREATE CSV FROM DATAFRAME**

```python
df.to_csv('complete-cryptojack-dataset.csv')
```
